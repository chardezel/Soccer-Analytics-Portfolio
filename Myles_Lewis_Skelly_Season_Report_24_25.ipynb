```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set(style="whitegrid")
plt.rcParams.update({'font.size': 12})

# 1. Consistency Over Time
months = ['Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar', 'Apr']
ratings = [6.6, 6.6, 6.6, 6.8, 7.0, 6.6, 6.7, 7.1, 6.9]

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(months, ratings, marker='o', linestyle='-', color='darkblue', linewidth=2)
ax.set_title('üîÅ Myles Lewis-Skelly: Monthly Match Ratings (2024/25)')
ax.set_xlabel('Month')
ax.set_ylabel('Avg Match Rating')
ax.set_ylim(6.4, 7.3)
ax.grid(True)
plt.tight_layout()
plt.show()

# 2. Advanced Metrics Bar Chart
metrics = ['Non-Penalty Goals', 'Pass Completion %', 'Progressive Passes', 'Shot-Creating Actions']
per_90 = [0.12, 91.9, 5.07, 1.27]
percentiles = [87, 99, 82, 9]

df = pd.DataFrame({'Metric': metrics, 'Per 90': per_90, 'Percentile': percentiles})

fig, ax1 = plt.subplots(figsize=(10, 5))
bar1 = sns.barplot(x='Per 90', y='Metric', data=df, color='skyblue', label='Per 90')

ax2 = ax1.twiny()
bar2 = sns.barplot(x='Percentile', y='Metric', data=df, color='orange', alpha=0.4, label='Percentile', ax=ax2)

for i, (p90, pct) in enumerate(zip(per_90, percentiles)):
    ax1.text(p90 + 0.15, i, f"{p90}", va='center')
    ax2.text(pct + 1, i, f"{pct}%", va='center')

ax1.set_title('üìä Advanced Stats ‚Äì Per 90 & Percentile vs. Fullbacks')
ax1.set_xlabel('Per 90 Value')
ax2.set_xlabel('Percentile')
ax1.legend(loc='lower right')
plt.tight_layout()
plt.show()

# 3. Radar Chart
labels = np.array(metrics)
num_vars = len(labels)
values_per_90 = per_90 + [per_90[0]]
values_percentile = percentiles + [percentiles[0]]
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))
ax.plot(angles, values_per_90, color='skyblue', linewidth=2, label='Per 90')
ax.fill(angles, values_per_90, color='skyblue', alpha=0.3)
ax.plot(angles, values_percentile, color='orange', linewidth=2, linestyle='dashed', label='Percentile')
ax.fill(angles, values_percentile, color='orange', alpha=0.2)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_title("üï∏Ô∏è Radar Profile ‚Äì Myles Lewis-Skelly")
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
