```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.offsetbox as offsetbox

stats = {
    "PSxG - GA": 0.02,
    "Goals Against": 1.49,
    "Save %": 65.7,
    "PSxG/SoT": 0.31,
    "Penalty Save %": 33.3,
    "Clean Sheet %": 24.2,
    "Touches": 33.89,
    "Launch %": 39.4,
    "Goal Kicks": 6.87,
    "GK Avg Length": 51.1,
    "Crosses Stopped %": 5.6,
    "Def. Actions Outside Box": 1.25,
    "Avg. Def. Action Distance": 15.5
}

df = pd.DataFrame(list(stats.items()), columns=["Statistic", "Per 90"])
df = df.sort_values("Per 90", ascending=True)

plt.figure(figsize=(12, 8))
bars = plt.barh(df["Statistic"], df["Per 90"], color="#FF8200", edgecolor="#002147")

plt.title(" Giorgi Mamardashvili  Valencia CF (2024-25 GK Report)", fontsize=16, color="#002147")
plt.xlabel("Per 90 Stats", fontsize=12, color="#002147")
plt.xticks(color="#002147")
plt.yticks(color="#002147")
plt.grid(axis='x', linestyle='--', alpha=0.3, color="#cccccc")

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.7, bar.get_y() + bar.get_height()/2,
             f"{width:.2f}", va='center', fontsize=9, color="#002147")


    logo = mpimg.imread("valencia_logo.png")
    imagebox = offsetbox.OffsetImage(logo, zoom=0.2)
    ab = offsetbox.AnnotationBbox(imagebox, (0.90, 0.70), xycoords='axes fraction', frameon=False)
    plt.gca().add_artist(ab)


plt.tight_layout()
plt.savefig("mamardashvili_valencia_report.png", dpi=300, bbox_inches="tight")
plt.show()

```
